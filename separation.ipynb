```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from functions import data_separation, plotting_pieces, plot_filtered_and_original

```

```python
all_sheets = pd.read_excel('MINDRAY Synovent E5.xlsx', sheet_name=None)
```

```python
col_medicine = []
lst_data_medicine = []
for i, sheet in enumerate(list(all_sheets.keys())):
    if i <= 3:
        continue
    df = all_sheets[sheet]
    col = df.columns.to_list()
    for col in col:
        is_all_numeric = pd.to_numeric(df[col], errors='coerce').notna().all()
        if is_all_numeric:
            data = df[col]
            data = np.array(data[~np.isnan(data)], dtype=float)
            lst_data_medicine.append({'name': col, 'data': data, 'num_points': 0})
            col_medicine.append(col)
data_df = pd.DataFrame(lst_data_medicine)
data_df.reset_index(inplace=True)
data_df
```

```python
num_points_df = ['Дата',
 ('Pпик(смH2O)', 2),
 'MV(л/мин)',
 'TVe(мл)',
 ('fобщ(дых/мин)', 2),
 'TVi(мл)',
 'fспонт(дых/мин)',
 'WOB(Дж/мин)',
 ('Pплато(смH2O)', 2),
 'Pсред(смH2O)',
 'TVe спонт(мл)',
 'TVe/ИМТ(мл/кг)',
 'Mvспонт(л/мин)',
 'MVутеч(л/мин)',
 ('fпринуд(дых/мин)', 2),
 'Дата',
 'Дата',
 ('Pпик(смH2O)', 2),
 'MV(л/мин)',
 'TVe(мл)',
 ('fобщ(дых/мин)', 3),
 'TVi(мл)',
 'fспонт(дых/мин)',
 'WOB(Дж/мин)',
 ('Pплато(смH2O)', 2),
 'Pсред(смH2O)',
 'TVe спонт(мл)',
 'TVe/ИМТ(мл/кг)',
 'Mvспонт(л/мин)',
 'MVутеч(л/мин)',
 'fпринуд(дых/мин)',
 'Дата',
 'O2%(%)',
 'Pвыс(смH2O)',
 'Pниз(смH2O)',
 'f(дых/мин)',
 'Tinsp(с)',
 'Tslope(с)',
 'dPsupp(смH2O)',
 'F-триг.(л/мин)',
 'Выдох%(%)',
 'dPапноэ(смH2O)',
 'fапноэ(дых/мин)',
 'Tinsp при апноэ(с)',
 'Внут.диам.труб.(мм)',
 'Компенс.(%)']
num_points_df = pd.DataFrame(num_points_df, columns=['flag'])
num_points_df.reset_index(inplace=True)
```

```python
res_df = pd.merge(data_df, num_points_df, on='index')
res_df['is_tuple'] = res_df['flag'].apply(lambda x: isinstance(x, tuple))
res_df = res_df.loc[res_df['is_tuple']]
res_df
```

```python
for index, row in res_df.iterrows():
    data = row['data']
    result = data_separation(data, row['flag'][1])
    y_min, y_max, x_min, x_max = np.min(data), np.max(data), 0, len(data)
    # save_total(data, result, y_min, y_max, x_min, x_max, len(data), row['name'])
    

    total_data = []
    for i in range(len(result) - 1):
        x_axis = np.arange(result[i], result[i + 1])
        name_graph = f"{row['name']} участок {i + 1}"
        temp = plotting_pieces(x_axis, data, y_min, y_max, x_min, x_max, len(data), name_graph)
        total_data.append(temp)
    name = f"{row['name']} участок общий"
    plot_filtered_and_original(result, data, total_data, name)
    
    
```
